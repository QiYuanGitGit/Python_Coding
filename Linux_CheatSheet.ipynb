#Security: Managing File Permissions and Ownership (**About Linux**)

Linux is a multi-user operating system.

Three possible levels of file ownership in Linux: user, group, and other.

Only an official owner of a file is allowed to change its permissions, can decide who can read the file, write to it, or execute it.

`rw- r-- r--`:
- Owner: rw-  read, write
- Group: r--  read only
- Others: r--  read only

###Viewing file permissions

`$ echo "Who can read this file?" > my_new_file
$ more my_new_file
Who can read this file?
$ ls -l my_new_file
-rw-r--r-- 1 theia users 25 Dec 22 17:47 x`



We echoed the string "Who can read this file?" into a new file called my_new_file. The next line uses the more command to print the contents of the new file. Finally, the ls command with the -l option displays the file's (default) permissions: rw-r--r--

First three characters (rw-) define the user permissions, the next three (r--) the group permissions, and the final three (r--) the other permissions.

So you, being the user, have the permission rw-, which means you have read and write permissions by default, but do not have execution permissions. Otherwise, there would be an x in place of the last -

Looking at the entire line, `rw-r--r--`, you can see that anyone can read the file, nobody can execute it, and you are the only user that can write to it.

###Making a file private

Revoke read permissions from your group and all other users by using the `chmod` command:

Eg.
`chmod go-r my_new_file`

`ls -l my_new_file`

`-rw------- 1 theia users 24 Dec 22 18:49 my_new_file`

Ensure successful modification by using the `ls -l` command

`go-r` to removing for the group (g) and others (o) the read (r) permission.

#Viewing and modifying file access permissions

To see the permissions for the file named `usdoi.txt` in your current directory:


Input: `ls -l usdoi.txt`

Output: `-rw-r--r-- 1 theia theia 8121 May 31 16:45 usdoi.txt`

Explanation: `-` indicates that `usdoi.txt` is a file, will see `d` instead of the `-` if **it were a directory**

Q1: revokes read permissions for all users (user, group, and other) on the file?

- `chmod -r usdoi.txt`          

Q2: Verify the changed permissions?

- `ls -l usdoi.txt`

Q3: To grant read access to all users on usdoi.txt?

- `chmod +r usdoi.txt`     

Q4: Remove the read permission only for 'other' category?

- `chmod o-r usdoi.txt`
