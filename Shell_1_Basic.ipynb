#**Pipes**

Pipes are commands in Linux which allow you to use the output of one command as the input of another.

image.png

Pipes | use the following format:

[command 1] | [command 2] | [command 3] ... | [command n]

no limit to the number of times you can chain pipes in a row

##Example pipes 1 (.txt)

`$ cat pets.txt`

cat\

`$ sort pets.txt`\

###sort + unqi it:

`$ sort pets.txt | uniq`\

## `tr` (translate) - replaces characters in input text

tr [OPTIONS] [target characters] [replacement characters]

`tr` to replace all the `vowels` in a string with underscores `_`:\
`$ echo "Linux and shell scripting are awesome\!" | tr "aeiou" "_"`\
L_n_x _nd sh_ll scr_pt_ng _r_ _w_s_m_!

`tr` to replace all the consonants (any letter that is not a vowel) with an underscore `-`:\
`$ echo "Linux and shell scripting are awesome\!" | tr -c "aeiou" "_"`\
_i_u__a_____e______i__i___a_e_a_e_o_e_

##`cat`(uppercase) + `tr`(replace with)

`$ cat pets.txt | tr "[a-z]" "[A-Z]"`\
GOLDFISH\
DOG\
CAT\
PARROT\
DOG\
GOLDFISH\
GOLDFISH

## `sort` + `cat` + `tr`

`$ sort pets.txt | uniq | tr "[a-z]" "[A-Z]"`\
CAT\
DOG\
GOLDFISH\
PARROT

get this json file to get the current price of Bitcoin (BTC) in USD, by using grep command. Then, copy the output in a file and name it as `Bitcoinprice.txt`

  "coin": {\

    "rank": 1,\
    "price": 57907.78008618953,\
    "priceBtc": 1,\
    "volume": 48430621052.9856,\
    "marketCap": 1093175428640.1146,\
    "availableSupply": 18877868,\

      "https://blockchair.com/bitcoin/",\

      "https://btc.tokenview.com/"\

JSON field you want to grab here is "price": [numbers].[numbers]". Use `grep` command to extract it from the JSON text:\
`grep -oE "\"price\"\s*:\s*[0-9]*\.?[0-9]*"`

* `-o` tells `grep` to only return the matching portion
* `-E` tells `gre`p to be able to use extended regex symbols such as `?`
* `\"price\"` matches the string `"price"`
* `\s`* matches any number (including 0) of whitespace (\s) characters
* `:` matches `:`
* `[0-9]*` matches any number of digits (from 0 to 9)
* `\.?` optionally matches `a` .

#Bash Shell Feature

##**Metacharaters** `#`,`;`,`*`,`?`

### `#`or called Pound (Comment)

echo "Hello, world!"  # This is another comment

### `;` (To separate multiple commands on a single command line)

$ echo "Hello, "; echo "world!"\
Hello,\
world!

### `*` (used as a wildcard character to represent any sequence of characters, including none.)

`ls *.txt`

will matches any file in the current directory with a .txt extension.

### `?` (used as a wildcard character to represent any single character)

`ls file?.txt`

for replace of single character

## **Quoting** `\`, `" "`, `' '`

###Backslash `\`

If you have a file with spaces in its name, you can use backslashes followed by a space to handle those spaces literally:\
`touch file\ with\ space.txt`

###Double quotes `" "`

`$ echo "Hello $USER"`\
Hello `<username>`

###Single quotes `' '`

`$ echo 'Hello $USER'`\
Hello $USER

Instead of printing the value of `$USER`, single quotes cause the terminal to print the string "$USER"

##**Input/Output redirection**

image.png

`sort < data.txt` will sort the contents of the data.txt file.

##**Command Substitution**

Allows you to run command and use its output as a component of another command's argument:\
`$ here=$(pwd)`\
`$ cd path_to_some_other_directory`\
`$ cd $here`

^^For example, you could store the path to your current directory in a variable by applying command substitution on the pwd command, then move to another directory, and finally return to your original directory by invoking the cd command on the variable you stored.

##**Command Line Arguments**

The following command provides two arguments, arg1, and arg2:

`$ ./MyBashScript.sh arg1 arg2`
