# **Brief Introduction to Networking**

##Computer Network, Network Resource, Network Node

- A computer network is a set of computers that are able to communicate with each other and share resources provided by network nodes.

- A network resource is any object, such as a file or document, which can be identified by the network.

- A network node is any device that participates in a network.

##Hosts, Clients, and Servers

- A host is a special type of node in a computer network - it is a computer that can function as a server or a client on a network.

- A server is a host computer that is able to accept a connection from a client host and fulfill certain resource requests made by the client.

- Many hosts can perform either role, acting as both client and server.



##Packets and Pings

- `Network packet` - a formatted chunk of data that can be transmitted over a network.

- `ping` - works by sending special 'echo request' packets to a host and waiting for a response from the host.

##URLs and IP Adresses

- IP or "Internet Protocol" - the format of data transmitted over the internet or a local network.

- An IP address - a code used to uniquely identify any host on a network.

- A URL or a web address, or Uniform Resource Locator - identifies a web resource and enables access to that resource.

#**Archiving and Compression of file and folder**

##(`tar`)Pack multiple files and directories into a single archive file

image.png

`tar -cvf bin.tar /bin`, create new archive directory '/bin', archive to a single file 'bin.tar'.

`t` in `tar -tvf bin.tar` to see the list of files in the archive.

`x` in `tar -xvf bin.tar` to untar the archive or extract files from the archive.

`ls -l` to verify that the folder bin is extracted.

##(`zip`)Compress to ZIP files

`zip config.zip /etc/*.conf` to to compress files, create zip file named 'config.zip', consists all the files with extension '.conf', in directory /etc.

`-r` to zip an entire directory.

`-y` to prevent symbolic links from being followed recursively.

##(`unzip`)Extract ZIP file

`unzip` to extract files.

`unzip -l config.zip` to list the files of the archive 'config.zip'

`unzip -o bin.zip`  to force overwrite in case you run the command more than once.
