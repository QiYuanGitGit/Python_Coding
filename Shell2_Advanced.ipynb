#**Conditionals** (if-then-else syntax)

if [ condition ]\
then\
    statement_block_1  
else\
    statement_block_2  
fi

If the `condition` is `true`, then Bash executes the statements in `statement_block_1` before exiting the conditional block of code. After exiting, it will continue to run any commands after the closing `fi`.

If the `condition` is `false`, Bash instead runs the statements in `statement_block_2` under the else line.

Tips:

* put spaces around your condition within the square brackets `[ ]`
* `if` condition block must be paired with a `fi`
* `else` is optional to put, w/o `else` block, nothing happens if `false`

Condition is checking whether the number of command-line arguments read by some Bash script, $#, is equal to 2:

if [[ $# == 2 ]]\
then\
  echo "number of arguments is equal to 2"\
else\
  echo "number of arguments is not equal to 2"\
fi

! double square brackets is syntax required for making integer comparisons in the condition [[ $# == 2 ]].

##Multiple conditions `&&` and `||`

### `&&` and

if [ condition1 ] && [ condition2 ]\
then\
    echo "conditions 1 and 2 are both true"
else\
    echo "one or both conditions are false"
fi

###`||` or

if [ condition1 ] || [ condition2 ]\
then\
    echo "conditions 1 or 2 are true"\
else\
    echo "both conditions are false"\
fi

#`for` loop

`for` loop requires a `; do` component in order to cycle through the loop. Additionally, you need to terminate the `for` loop block with a `done` statement.

for i in ${!my_array[@]}; do

  echo ${my_array[$i]}
  
done

###`for` with N iterations

N=6\
for (( i=0; i<=$N; i++ )) ; do

  echo $i
  
done

`sum=$(($n1+$n2))`\

###2.2 Bash script Add logic

`if [ $sum -lt $product ]`\
then\
   echo "The sum is less than the product."
`elif [ $sum -eq $product ]`\
then\
   echo "The sum is equal to the product."\
`elif [ $sum -gt $product ]`\
then\
   `echo "The sum is greater than the product."`\
fi

`-lt` is less than

`eq` is equal

`gt` is greater than

`if` is closed the block

| Operator | Meaning | Example |  
|----------|---------|---------|  
| `-eq` | Equal to | `[ $a -eq $b ]` |  
| `-ne` | Not equal to | `[ $a -ne $b ]` |  
| `-lt` | Less than | `[ $a -lt $b ]` |  
| `-le` | Less than or equal | `[ $a -le $b ]` |  
| `-gt` | Greater than | `[ $a -gt $b ]` |  
| `-ge` | Greater than or equal | `[ $a -ge $b ]` |
